```python
import numpy as np
import pandas as pd 
from datetime import timedelta
import datetime as dt
import matplotlib.pyplot as plt
df=pd.read_csv('https://raw.githubusercontent.com//bipulshahi//Dataset//main//Walmart.csv ')
```

```python
#Total number of orders made from each state

df.groupby('State')['Order ID'].agg(['count']).sort_values('count' , ascending = False)
```

```python
#Total quantities sold from each category

df.groupby('Category')['Quantity'].agg(['sum']).sort_values('sum' , ascending = False)
```

```python
#Number of orders of each category from each state

df.groupby(['State','Category'])['Order ID'].count().unstack()
```

```python
#Total quantity sold of each category from each state

df.groupby(['State','Category'])['Quantity'].sum().unstack()
```

```python
#Average of sales and profit made by each category

df.groupby('Category')[['Sales','Profit']].mean()
```

```python
#Total of sales and profit made by each category

df.groupby('Category')[['Sales','Profit']].sum()
```

```python
#Processing Time of each order -> difference between ship date

df['Order_date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['ship_date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')
```

```python
df['O_day'] = df['Order_date'].dt.day
df['O_month'] = df['Order_date'].dt.month
df['O_year'] = df['Order_date'].dt.year
```

```python
df['processing_time'] = (df['ship_date'] - df['Order_date']).map(lambda x: x.days)
df['processing_time']
```

```python
#Total sales & profit made in each year

df.groupby('O_year',as_index=False)[['Sales','Profit']].sum()
```
